```python
news_df["text"] = news_df["title"] + news_df["text"]
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from transformers import DataCollatorWithPadding, AutoTokenizer
from transformers import AutoModelForSequenceClassification, TrainingArguments, Trainer
from transformers import EarlyStoppingCallback, IntervalStrategy
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score
from datasets import load_metric

class NewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item["labels"] = torch.tensor([self.labels[idx]])
        return item

    def __len__(self):
        return len(self.labels)


def compute_metrics(p):    
  pred, labels = p
  pred = np.argmax(pred, axis=1)
  accuracy = accuracy_score(y_true=labels, y_pred=pred)
  recall = recall_score(y_true=labels, y_pred=pred)
  precision = precision_score(y_true=labels, y_pred=pred)
  f1 = f1_score(y_true=labels, y_pred=pred)    
  return {"accuracy": accuracy, "precision": precision, "recall": recall, "f1": f1}


def to_dict(text,labels):
  dictionary = []
  for txt, y in zip(text, labels):
    dictionary.append({
        "text": txt,
        "label": y
    })
  return dictionary
```

```python
train_X, test_X, train_y, test_y = train_test_split(news_df["text"], news_df["label"], stratify=news_df["label"], test_size = 0.2, random_state = 10)
train_X, dev_X, train_y, dev_y = train_test_split(train_X, train_y, stratify=train_y, test_size = 0.25, random_state = 10) # 0.25 x 0.8 = 0.2

test_X.to_csv("News_dataset/testX.csv", index=False)
test_y.to_csv("News_dataset/testY.csv", index=False)

train_spans = to_dict(train_X, train_y)
dev_spans = to_dict(dev_X, dev_y)
test_spans = to_dict(test_X, test_y)

train_spans_txt = [s['text'] for s in train_spans] 
dev_spans_txt = [s['text'] for s in dev_spans] 
test_spans_txt = [s['text'] for s in test_spans]

train_y = [s['label'] for s in train_spans]
dev_y = [s['label'] for s in dev_spans]
test_y = [s['label'] for s in test_spans]
```

```python
# max sequence length for each document/sentence sample
max_length = 512

# load the tokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')

train_encodings = tokenizer(train_spans_txt, truncation=True, padding=True, max_length=max_length)
dev_encodings = tokenizer(dev_spans_txt, truncation=True, padding=True, max_length=max_length)
test_encodings = tokenizer(test_spans_txt, truncation=True, padding=True, max_length=max_length)
```

```python
# convert our tokenized data into a torch Dataset
train_dataset = NewsDataset(train_encodings, train_y)
dev_dataset = NewsDataset(dev_encodings, dev_y)
test_dataset = NewsDataset(test_encodings, test_y)

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased", num_labels=2)

training_args = TrainingArguments(  
    learning_rate=5e-5, # default
    weight_decay=0.01,
    num_train_epochs=3,
    evaluation_strategy=IntervalStrategy.STEPS, # evaluation done and log every eval_steps
    eval_steps=200,
    logging_steps=200,            
    save_steps=400,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    output_dir="./results",
    logging_dir='./logs',   
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=dev_dataset,
    data_collator=data_collator,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
    callbacks = [EarlyStoppingCallback(early_stopping_patience=3)]) # Adopting early stopping. 
                                                                    # Bert needs much less than three epochs for the fake news detection task 
                                                                    # for high accuracy on dev set.
```

```python
trainer.train()
```

```python
trainer.evaluate()
```

```python
trainer.evaluate(test_dataset)
```
